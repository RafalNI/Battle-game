### Zadanie 1 Wektor [40]
Zaimplementuj klasę wektor:
(dodawanie, mnożenie przez skalar, odejmowanie, wyznaczenie długości, iloczyn skalarny wektorów, czy wektory sa równe, czy wektory są prostopadłe, czy wektory są równoległe)

```python
class Vector():
    
    def __init__(self, Vector):
        self.set_Vector(Vector)
        
    def set_Vector(self, Vector):
        self.Vector = Vector
        
    def get_Vector(self):
        return self.Vector
    
    def plus(self, vector_add):
        self.Vector[0] = self.get_Vector()[0] + vector_add[0]
        self.Vector[1] = self.get_Vector()[1] + vector_add[1]
        
    def minus(self, vector_sub):
        self.Vector[0] = self.get_Vector()[0] - vector_sub[0]
        self.Vector[1] = self.get_Vector()[1] - vector_sub[1]
        #self.Vector = self.plus(self, -vector_sub)
    
    def length(self):
        from math import sqrt
        return sqrt((self.get_Vector()[0])**2 + (self.get_Vector()[1])**2)
    
    def multi(self, vector_mul):
        self.Vector[0] = self.get_Vector()[0] * vector_mul[0]
        self.Vector[1] = self.get_Vector()[1] * vector_mul[1]
        
    def equal(self, vector_eq):
        if self.get_Vector() > vector_eq:
            return f'({self.get_Vector()}) is greater than ({vector_eq})'
        elif self.get_Vector() == vector_eq:
            return f'({self.get_Vector()}) is equal to ({vector_eq})'
        else:
            return f'({self.get_Vector()}) is lesser than ({vector_eq})'
    
    def parallel(self, other):
        if self.get_Vector()[0]*other[1] - self.get_Vector()[1]*other[0] == 0:
            return f"Vector {self.get_Vector()} is parallel to {other}"
        else:
            return f"Vector {self.get_Vector()} is NOT parallel to {other}"
    
    def perpendicular(self, other):
        if self.get_Vector()[0]*other[0] + self.get_Vector()[1]*other[1] == 0:
            return f"Vector {self.get_Vector()} is perpendicular to {other}"
        else:
            return f"Vector {self.get_Vector()} is NOT perpendicular to {other}"
    
    def __str__(self):
        return f"Vector is equal to: {self.get_Vector()} and is {self.length():.2f} long."
    
    def __repr__(self):
        return self.__str__()
    
Wektor1 = Vector([1,4])
print(Wektor1)
Wektor1.minus([0,1])
print(Wektor1)
Wektor1.plus([0,3])
print(Wektor1)
print(Wektor1.equal([1,5]))
print(Wektor1.equal([1,6]))
print(Wektor1.equal([1,7]))
Wektor1.multi([2,2])
print(Wektor1.parallel([2,12]))
print(Wektor1.perpendicular([2,12]))
print(Wektor1)
```

```python
class GRA:
    '''
    Gra dziala w taki sposob, że walczy sie do czasu aż nie wyskoczy error.
    W przypisie errora powinno być napisane który z graczy wygrywa.
    
    Aby zacząć grać wystarczy podać w konsoli komendę: 'cokolwiek = GRA('pojazd1', 'pojazd2')'
    
    Pojazdy do wyboru:
    - Tank
    - BattleShip
    
    Wystarczy czytać co gra pokazuję i stosować się do poleceń i powinno być OK!
    
    Have fun!
    '''
      
    def __init__(self, P1, P2):
        self.set_licznik_tur()
        self.P1 = P1
        self.P2 = P2
        self.tura_loop()
        
    def set_licznik_tur(self):
        self.licznik_tur = 1
        
    def get_licznik_tur(self):
        return self.licznik_tur
        
    def set_P1T(self, var = 5, miej = (0,0), zdrowie = 6):
        self.mocP1 = var
        self.miejsceP1 = miej
        self.zdrowieP1 = zdrowie
        
    def set_P1B(self, var = 15, miej = (-4,4), zdrowie = 12):
        self.mocP1 = var
        self.miejsceP1 = miej
        self.zdrowieP1 = zdrowie
        
    def set_P2T(self, var = 5, miej = (0,0), zdrowie = 6):
        self.mocP2 = var
        self.miejsceP2 = miej
        self.zdrowieP2 = zdrowie
        
    def set_P2B(self, var = 15, miej = (-4,4), zdrowie = 12):
        self.mocP2 = var
        self.miejsceP2 = miej
        self.zdrowieP2 = zdrowie
        
    def set_Player1(self, nazwa):
        if nazwa == 'Tank':
            if self.get_licznik_tur() == 1:
                self.Player1 = Tank("M3 General Lee")
            else:
                self.set_P1T(self.Player1.get_moc(), self.Player1.get_miejsce(), self.Player1.get_zdrowie())
                self.Player1 = Tank("M3 General Lee", self.mocP1, self.miejsceP1, self.zdrowieP1)
        elif nazwa == 'BattleShip':
            if self.get_licznik_tur() == 1:
                self.Player1 = BattleShip("Grom")
            else:
                self.set_P1B(self.Player1.get_moc(), self.Player1.get_miejsce(), self.Player1.get_zdrowie())
                self.Player1 = BattleShip("Grom", self.mocP1, self.miejsceP1, self.zdrowieP1)
        
    def get_Player1(self):
        return self.Player1
    
    def set_Player2(self, nazwa):
        if nazwa == 'Tank':
            if self.get_licznik_tur() == 1:
                self.Player2 = Tank("M3 General Lee")
            else:
                self.set_P2T(self.Player2.get_moc(), self.Player2.get_miejsce(), self.Player2.get_zdrowie())
                self.Player2 = Tank("M3 General Lee", self.mocP2, self.miejsceP2, self.zdrowieP2)
        elif nazwa == 'BattleShip':
            if self.get_licznik_tur() == 1:
                self.Player2 = BattleShip("Grom")
            else:
                self.set_P2B(self.Player2.get_moc(), self.Player2.get_miejsce(), self.Player2.get_zdrowie())
                self.Player2 = BattleShip("Grom", self.mocP2, self.miejsceP2, self.zdrowieP2)
        
    def get_Player2(self):
        return self.Player2
    
    def tura_loop(self):
        while True:
            self.turaP1()
            self.turaP2()
            self.licznik_tur += 1
    
    def turaP1(self):
        self.set_Player1(self.P1)
        print(f"Tura {self.licznik_tur}:")
        while self.get_Player1().get_punkty_ruchu() > 0:
            self.actualPlayer = self.get_Player1()
            self.sprawdz_zdrowie()
            self.sprawdz_moc()
            print(self.actualPlayer)
            self.dzialania()
            self.sprawdz_zdrowie()
            self.sprawdz_moc()
        print(self.actualPlayer)
        print("Koniec tury\nWyczerpałeś punkty ruchu\n\nRuch gracza 2:")
          
    def turaP2(self):
        self.set_Player2(self.P2)
        while self.get_Player2().get_punkty_ruchu() > 0:
            self.actualPlayer = self.get_Player2()
            self.sprawdz_zdrowie()
            self.sprawdz_moc()
            print(self.actualPlayer)
            self.dzialania()
            self.sprawdz_zdrowie()
            self.sprawdz_moc()
        print(self.actualPlayer)
        print("Koniec tury\nWyczerpałeś punkty ruchu\n\n")
              
    def dzialania(self):
        print("Jakie działanie chcesz wykonac? (ruch poziomy, ruch pionowy, atak)")
        buf = input()
        if buf == 'ruch poziomy':
            self.poziomy()
        elif buf == 'ruch pionowy':
            self.pionowy()
        elif buf == 'atak':
            self.atakuj()
    
    def poziomy(self):
        print("Podaj o ile chcesz się przesunac poziomo")
        x = input()
        self.actualPlayer.ruch_poziomy(int(x))
    
    def pionowy(self):
        print("Podaj o ile chcesz się przesunac pionowo")
        y = input()
        self.actualPlayer.ruch_pionowy(int(y))
    
    def atakuj(self):
        print("Podaj wspolrzedna X swojego celu")
        x = input()
        print("Podaj wspolrzedna Y swojego celu")
        y = input()
        self.actualPlayer.atak((int(x), int(y)))
        xy = [int(x), int(y)]
        if self.actualPlayer == self.Player1:
            if self.get_Player2().get_miejsce() == xy:
                self.Player2.odejmij_zdrowie(int(self.actualPlayer.sila()))
                self.sprawdz_zdrowie()
                print(f"Pozostałe punkty zdrowia przeciwnika: {self.Player2.get_zdrowie()})\n")
        elif self.actualPlayer == self.Player2:
            if self.get_Player1().get_miejsce() == xy:
                self.Player1.odejmij_zdrowie(int(self.actualPlayer.sila()))
                self.sprawdz_zdrowie()
                print(f"Pozostałe punkty zdrowia przeciwnika: {self.Player1.get_zdrowie()})\n")
        
    def sprawdz_miejsce(self):
        '''Pojazdy nie moga stac na tym samym polu'''
        if self.Player1.get_miejsce() == self.Player2.get_miejsce():
            raise Exception ("Nie możesz ruszyć się na to pole, już tam stoi jakiś pojazd.")
            
    def sprawdz_zdrowie(self):
        if self.actualPlayer.get_zdrowie() <= 0:
            if self.actualPlayer == self.Player1:
                raise Exception ("Gracz 2 wygrywa eliminując przeciwnika!")
            elif self.actualPlayer == self.Player2:
                raise Exception ("Gracz 1 wygrywa eliminując przeciwnika!")
                
    def sprawdz_moc(self):
        if self.actualPlayer.get_moc() <= 0:
            if self.actualPlayer == self.get_Player1():
                raise Exception ("Gracz 1 wyczerpał swoją moc.\n\n ---Wygrywa gracz 2!")
            elif self.actualPlayer == self.get_Player2():
                raise Exception ("Gracz 2 wyczerpał swoją moc. \n\n ---Wygrywa gracz 1!")

### -------------------------------------- ###
    
gra1 = GRA('Tank', 'BattleShip')
```
